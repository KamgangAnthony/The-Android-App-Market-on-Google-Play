```python
import pandas as pd
```

```python
apps = pd.read_csv('datasets/apps.csv')
```

```python
apps.info()
```

```python
apps.head()
```

```python
chars_to_replace = ['+', ',']
```

```python
for char in chars_to_replace:
    apps['Installs'] = apps['Installs'].apply(lambda x: x.replace(char,''))
```

```python
apps['Installs'] = apps['Installs'].astype(int)
```

```python
apps.info()
```

```python
app_category_info = apps.groupby('Category').agg({'App':'count','Price':'mean','Rating':'mean'})
app_category_info
```

```python
app_category_info = app_category_info.rename(columns={'App':'Number of apps','Price':'Average price','Rating':'Average rating'})
```

```python
app_category_info = app_category_info.reset_index()
app_category_info
```

```python
reviews = pd.read_csv('datasets/user_reviews.csv')
```

```python
reviews.head()
```

```python
condition = (apps['Type'] == 'Free') & (apps['Category'] == 'FINANCE')
```

```python
free_finance_apps = apps[(condition)]
free_finance_apps.head()
```

```python
free_finance_apps_with_reviews = pd.merge(free_finance_apps, reviews, on='App')
free_finance_apps_with_reviews.head()
```

```python
user_feedback = free_finance_apps_with_reviews.groupby('App').agg({'Sentiment Score':'mean'})
user_feedback.head()
```

```python
user_feedback = user_feedback.sort_values(by='Sentiment Score', ascending=False).reset_index()
user_feedback.head()
```

```python
top_10_user_feedback = user_feedback[:10]
top_10_user_feedback
```
